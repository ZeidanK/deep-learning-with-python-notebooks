```python

# Loading the IMDB dataset
#******************************
from tensorflow.keras.datasets import imdb
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

train_data[0]

train_labels[0]

max([max(sequence) for sequence in train_data])


# Decoding reviews back to text
#******************************
word_index = imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
decoded_review = " ".join([reverse_word_index.get(i - 3, "?") for i in train_data[0]])
decoded_review

#Preparing the data
#Encoding the integer sequences via multi-hot encoding
#*****************************************************
import numpy as np
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        for j in sequence:
            results[i, j] = 1.
    return results
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

x_train[0]
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")

# Building your model
# Model definition
#******************************
from tensorflow import keras
from tensorflow.keras import layers
from keras import regularizers
model = keras.Sequential()
#model.add(layers.Dense(16, kernel_regularizer=regularizers.l1(0.0005), activation="relu"))
model.add(layers.Dense(16, activation="relu"))
#model.add(layers.Dropout(0.5))
#model.add(layers.Dense(16, kernel_regularizer=regularizers.l1(0.0005), activation="relu"))
model.add(layers.Dense(16, activation="relu"))
#model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation="sigmoid"))

# Compiling the model
#******************************
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

# Validating your approach
# Setting aside a validation set
#*******************************
x_val = x_train[:10000] # to be used for validation
partial_x_train = x_train[10000:] 
y_val = y_train[:10000] # to be used for validation
partial_y_train = y_train[10000:]


# Training your model
#*******************************
history = model.fit(partial_x_train,
                    partial_y_train,
                    epochs=30,
                    batch_size=512,
                    validation_data=(x_val, y_val))

history_dict = history.history
history_dict.keys()

# Plotting the training and validation loss
#******************************************
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict["loss"]
val_loss_values = history_dict["val_loss"]
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, "bo", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()


# Plotting the training and validation accuracy
#**********************************************
plt.clf()
acc = history_dict["accuracy"]
val_acc = history_dict["val_accuracy"]
plt.plot(epochs, acc, "bo", label="Training acc")
plt.plot(epochs, val_acc, "b", label="Validation acc")
plt.title("Training and validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# Retraining a model from scratch
#********************************
model = keras.Sequential([
    layers.Dense(16, activation="relu"),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
model.fit(x_train, y_train, epochs=4, batch_size=512)
results = model.evaluate(x_test, y_test)

results

# Using a trained model to generate predictions on new data
#**********************************************************
model.predict(x_test)

```

## Experiment 3: L2 Regularization - 16X16X1

**Description:** Apply L2 regularization (weight decay) to both hidden layers:
- **L2 penalty**: 0.001 (tested value after trying 0.0001 and 0.001)
- L2 regularization adds the sum of squared weights to the loss function
- Formula: `loss = original_loss + λ * Σ(weight²)`

**Purpose:** Penalize large weights to prevent the model from fitting too closely to training data.

**How it works:** L2 forces weights to be small and evenly distributed, preventing any single feature from dominating. This creates smoother, more generalizable decision boundaries.

**Expected Result:** Lower overfitting, slightly lower training accuracy but better validation accuracy.

```python
# Experiment 3: L2 Regularization
print("\n" + "=" * 60)
print("EXPERIMENT 3: L2 REGULARIZATION (λ=0.001)")
print("=" * 60)

# Build model with L2 regularization
# Tried: 0.0001 (too weak), 0.001 (good balance)
model_3 = keras.Sequential([
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), 
                 activation="relu", input_shape=(10000,)),
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.001), 
                 activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model_3.compile(optimizer="rmsprop",
                loss="binary_crossentropy",
                metrics=["accuracy"])

# Train model
history_3 = model_3.fit(partial_x_train,
                        partial_y_train,
                        epochs=30,
                        batch_size=512,
                        validation_data=(x_val, y_val),
                        verbose=0)

# Store results
all_histories["L2 Reg (16X16X1)"] = history_3.history["val_accuracy"]

print(f"L2 Regularization weight: 0.001")
print(f"Final Validation Accuracy: {history_3.history['val_accuracy'][-1]:.4f}")
print(f"Best Validation Accuracy: {max(history_3.history['val_accuracy']):.4f} at epoch {history_3.history['val_accuracy'].index(max(history_3.history['val_accuracy'])) + 1}")
print(f"Training Accuracy: {history_3.history['accuracy'][-1]:.4f}")
print(f"Overfitting Gap: {history_3.history['accuracy'][-1] - history_3.history['val_accuracy'][-1]:.4f}")
print(f"Comparison to Baseline: {history_3.history['val_accuracy'][-1] - baseline_val_acc[-1]:+.4f}")
```

## Experiment 4: L1 Regularization - 16X16X1

**Description:** Apply L1 regularization (Lasso) to both hidden layers:
- **L1 penalty**: 0.0005 (tested value after trying 0.0001 and 0.0005)
- L1 regularization adds the sum of absolute weights to the loss function
- Formula: `loss = original_loss + λ * Σ|weight|`

**Purpose:** Encourage sparsity - push less important weights to exactly zero.

**How it works:** L1 tends to eliminate weak connections entirely, creating sparse networks. This is useful for feature selection as it identifies which input features are most important.

**L1 vs L2:**
- L1: Creates sparse models, some weights → 0
- L2: Shrinks all weights but keeps them non-zero
- L1 better for feature selection, L2 better for general regularization

**Expected Result:** More sparse weight matrices, potential feature selection, reduced overfitting.

```python
# Experiment 4: L1 Regularization
print("\n" + "=" * 60)
print("EXPERIMENT 4: L1 REGULARIZATION (λ=0.0005)")
print("=" * 60)

# Build model with L1 regularization
# Tried: 0.0001 (too weak), 0.0005 (good balance)
model_4 = keras.Sequential([
    layers.Dense(16, kernel_regularizer=regularizers.l1(0.0005), 
                 activation="relu", input_shape=(10000,)),
    layers.Dense(16, kernel_regularizer=regularizers.l1(0.0005), 
                 activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

model_4.compile(optimizer="rmsprop",
                loss="binary_crossentropy",
                metrics=["accuracy"])

# Train model
history_4 = model_4.fit(partial_x_train,
                        partial_y_train,
                        epochs=30,
                        batch_size=512,
                        validation_data=(x_val, y_val),
                        verbose=0)

# Store results
all_histories["L1 Reg (16X16X1)"] = history_4.history["val_accuracy"]

print(f"L1 Regularization weight: 0.0005")
print(f"Final Validation Accuracy: {history_4.history['val_accuracy'][-1]:.4f}")
print(f"Best Validation Accuracy: {max(history_4.history['val_accuracy']):.4f} at epoch {history_4.history['val_accuracy'].index(max(history_4.history['val_accuracy'])) + 1}")
print(f"Training Accuracy: {history_4.history['accuracy'][-1]:.4f}")
print(f"Overfitting Gap: {history_4.history['accuracy'][-1] - history_4.history['val_accuracy'][-1]:.4f}")
print(f"Comparison to Baseline: {history_4.history['val_accuracy'][-1] - baseline_val_acc[-1]:+.4f}")
```

## Experiment 5: Dropout - 16X16X1

**Description:** Apply Dropout regularization:
- **Dropout rate**: 0.5 (tested rates: 0.3, 0.5)
- Dropout randomly sets 50% of outputs to zero during training
- Applied after each hidden layer

**Purpose:** Prevent co-adaptation of neurons by forcing the network to learn redundant representations.

**How it works:** 
- During training: Randomly "drops" neurons with probability p (here 0.5)
- During inference: Uses all neurons but scales outputs by (1-p)
- Forces network to not rely on any single neuron
- Creates ensemble-like effect with many "sub-networks"

**Why Dropout works:**
- Prevents complex co-adaptations where neurons rely on specific other neurons
- Forces each neuron to learn useful features independently
- Provides strong regularization without modifying loss function

**Expected Result:** Significant reduction in overfitting, more robust model.

```python
# Experiment 5: Dropout
print("\n" + "=" * 60)
print("EXPERIMENT 5: DROPOUT (rate=0.5)")
print("=" * 60)

# Build model with Dropout
# Tried: 0.3 (moderate), 0.5 (aggressive)
model_5 = keras.Sequential([
    layers.Dense(16, activation="relu", input_shape=(10000,)),
    layers.Dropout(0.5),
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")
])

model_5.compile(optimizer="rmsprop",
                loss="binary_crossentropy",
                metrics=["accuracy"])

# Train model
history_5 = model_5.fit(partial_x_train,
                        partial_y_train,
                        epochs=30,
                        batch_size=512,
                        validation_data=(x_val, y_val),
                        verbose=0)

# Store results
all_histories["Dropout 0.5 (16X16X1)"] = history_5.history["val_accuracy"]

print(f"Dropout rate: 0.5 (50% of neurons dropped during training)")
print(f"Final Validation Accuracy: {history_5.history['val_accuracy'][-1]:.4f}")
print(f"Best Validation Accuracy: {max(history_5.history['val_accuracy']):.4f} at epoch {history_5.history['val_accuracy'].index(max(history_5.history['val_accuracy'])) + 1}")
print(f"Training Accuracy: {history_5.history['accuracy'][-1]:.4f}")
print(f"Overfitting Gap: {history_5.history['accuracy'][-1] - history_5.history['val_accuracy'][-1]:.4f}")
print(f"Comparison to Baseline: {history_5.history['val_accuracy'][-1] - baseline_val_acc[-1]:+.4f}")
```

**Purpose:** Test if reducing network capacity is sufficient to prevent overfitting.

- Smaller networks have less capacity to memorize training data
- Fewer parameters = less ability to fit noise
- Acts as implicit regularization through capacity constraint
- Trade-off: May underfit if too small

**Architecture Regularization vs Explicit Regularization:**
- Architecture: Limits what the model CAN learn
- Explicit (L1/L2/Dropout): Limits what the model DOES learn

**Expected Result:** Less overfitting due to reduced capacity, but possibly lower overall performance if network is too small.

## מידע נוסף - Additional Information

### Regularization Techniques Summary:

**1. L2 Regularization (Ridge)**
- Pros: Smooth, all features contribute, prevents extreme weights
- Cons: Doesn't eliminate features, all weights shrunk equally
- Best for: General regularization, when all features are relevant

**2. L1 Regularization (Lasso)**
- Pros: Feature selection, sparse models, interpretable
- Cons: Can be unstable, arbitrary feature selection when features are correlated
- Best for: Feature selection, high-dimensional data

**3. Dropout**
- Pros: Very effective, ensemble effect, no hyperparameter tuning of weights
- Cons: Slower training, needs more epochs, not applicable to all layers
- Best for: Deep networks, when you have enough data

**4. Architecture Regularization (Smaller Networks)**
- Pros: Faster training, less memory, implicit regularization
- Cons: Risk of underfitting, less capacity for complex patterns
- Best for: When you know the problem is relatively simple

### Tips for Further Exploration:
- Try combining L2 + Dropout
- Experiment with different network architectures (32X16X8X1)
- Try different optimizers (Adam, SGD with momentum)
- Use early stopping based on validation loss
- Try batch normalization as another form of regularization
